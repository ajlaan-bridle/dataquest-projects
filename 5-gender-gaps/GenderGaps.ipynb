# Visualising Gender Gaps in College Degrees

The aim of this project was to get some practice in formatting plots for storytelling purposes, particularly when it comes to cleaning up non-data elements to maximise the value of data visualisations. The end result is a grid of plots expressing a high level overview of gender gaps across a range of college degree subjects, over a number of decades.

The following elements are demonstrated using Matplotlib:
- Creating a 6x3 subplot grid to plot gender gaps for each subject in the data set.
- Generating line plots with consistent, colour-blind appropriate, colouring.
- Annotating plots with text in place of a bulky legend.
- De-cluttering the axes for each plot and adding a horizontal grey line in each plot to indicate gender balance at 50%.
- Exporting the plots as an image for downstream reporting use.

The [data](http://www.randalolson.com/wp-content/uploads/percent-bachelors-degrees-women-usa.csv) for this project comes from the personal website of Randal Olson, a data scientist at the University of Pennsylvania. It is a cleaned up version of data that is released annually by the Department of Education Statistics. The data set is a table of percentages of bachelor's degrees awarded to women for a range of subjects between the years 1970-2012.

```python
# Jupyter magic and importing pandas & pyplot modules
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline

# import data and some other setup variables
# degree subjects are categorised into columns of STEM, liberal arts, and other subjects
women_degrees = pd.read_csv('./percent-bachelors-degrees-women-usa.csv')
cb_dark_blue = (0/255,107/255,164/255)  # colour blind friendly dark blue
cb_orange = (255/255, 128/255, 14/255)  # colour blind friendly orange
stem_cats = ['Engineering', 'Computer Science', 'Psychology', 'Biology', 'Physical Sciences', 'Math and Statistics']
lib_arts_cats = ['Foreign Languages', 'English', 'Communications and Journalism', 'Art and Performance', 'Social Sciences and History']
other_cats = ['Health Professions', 'Public Administration', 'Education', 'Agriculture','Business', 'Architecture']

# create figure and axes
fig = plt.figure(figsize=(9, 18))

# generate stem_cats plots in figure
for sp in range(0, len(stem_cats)):
    ax = fig.add_subplot(6,3,(3*sp+1))
    ax.plot(
        women_degrees['Year'],          # x series
        women_degrees[stem_cats[sp]],   # y series
        c=cb_dark_blue,                 # series colour
        label='Women',                  # series label
        linewidth=3                     # line width
        )
    ax.plot(
        women_degrees['Year'], 
        100-women_degrees[stem_cats[sp]],   # men percentage is 100-women
        c=cb_orange, 
        label='Men', 
        linewidth=3)
    ax.spines["right"].set_visible(False)    
    ax.spines["left"].set_visible(False)
    ax.spines["top"].set_visible(False)    
    ax.spines["bottom"].set_visible(False)
    ax.set_xlim(1968, 2011)
    ax.set_ylim(0,100)
    ax.set_title(stem_cats[sp])
    ax.tick_params(bottom=False, top=False, left=False, right=False)
    ax.set_yticks([0, 50, 100])
    ax.axhline(50, c=(171/255, 171/255, 171/255), alpha=0.3)    # add translucent horizontal 50% line

    # remove bottom label on all but last one
    if sp < (len(stem_cats)-1):
        ax.tick_params(labelbottom=False)
    
    # label lines for first and last plots => legend not required
    if sp == 0:
        ax.text(2005, 87, 'Men')
        ax.text(2002, 8, 'Women')
    elif sp == 5:
        ax.text(2005, 62, 'Men')
        ax.text(2001, 35, 'Women')

# generate lib_arts_cats plots in figure
for sp in range(0, len(lib_arts_cats)):
    ax = fig.add_subplot(6,3,(3*sp+2))
    ax.plot(
        women_degrees['Year'],              # x series
        women_degrees[lib_arts_cats[sp]],   # y series
        c=cb_dark_blue,                     # series colour
        label='Women',                      # series label
        linewidth=3                         # line width
        )
    ax.plot(
        women_degrees['Year'], 
        100-women_degrees[lib_arts_cats[sp]],   # men percentage is 100-women
        c=cb_orange, 
        label='Men', 
        linewidth=3)
    ax.spines["right"].set_visible(False)    
    ax.spines["left"].set_visible(False)
    ax.spines["top"].set_visible(False)    
    ax.spines["bottom"].set_visible(False)
    ax.set_xlim(1968, 2011)
    ax.set_ylim(0,100)
    ax.set_title(lib_arts_cats[sp])
    ax.tick_params(bottom=False, top=False, left=False, right=False)
    ax.set_yticks([0, 100])
    ax.axhline(50, c=(171/255, 171/255, 171/255), alpha=0.3)

    # remove bottom label on all but last one
    if sp < (len(lib_arts_cats)-1):
        ax.tick_params(labelbottom=False)
    
    # label lines for first and last plots => legend not required
    if sp == 0:
        ax.text(2005, 20, 'Men')
        ax.text(2001, 75, 'Women')
    elif sp == 3:
        ax.text(2005, 30, 'Men')
        ax.text(2001, 65, 'Women')

# generate other_cats plots in figure
for sp in range(0, len(other_cats)):
    ax = fig.add_subplot(6,3,(3*sp+3))
    ax.plot(
        women_degrees['Year'],              # x series
        women_degrees[other_cats[sp]],   # y series
        c=cb_dark_blue,                     # series colour
        label='Women',                      # series label
        linewidth=3                         # line width
        )
    ax.plot(
        women_degrees['Year'], 
        100-women_degrees[other_cats[sp]],   # men percentage is 100-women
        c=cb_orange, 
        label='Men', 
        linewidth=3)
    ax.spines["right"].set_visible(False)    
    ax.spines["left"].set_visible(False)
    ax.spines["top"].set_visible(False)    
    ax.spines["bottom"].set_visible(False)
    ax.set_xlim(1968, 2011)
    ax.set_ylim(0,100)
    ax.set_title(other_cats[sp])
    ax.tick_params(bottom=False, top=False, left=False, right=False)
    ax.set_yticks([0, 100])
    ax.axhline(50, c=(171/255, 171/255, 171/255), alpha=0.3)

    # remove bottom label on all but last one
    if sp < (len(other_cats)-1):
        ax.tick_params(labelbottom=False)
    
    # label lines for first and last plots => legend not required
    if sp == 0:
        ax.text(2005, 5, 'Men')
        ax.text(2001, 90, 'Women')
    elif sp == 5:
        ax.text(2005, 65, 'Men')
        ax.text(2001, 30, 'Women')

# export plots
plt.savefig('./gender_degrees.png')

# display plots
plt.show()
```
